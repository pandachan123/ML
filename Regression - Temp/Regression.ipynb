```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
data = pd.read_csv('/content/temperatures - temperatures.csv')
```

```python
data.info()
```

```python
data.shape
```

```python
x=data[['YEAR','JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']]
y=data[['ANNUAL']]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
model = LinearRegression();
```

```python
model.fit(x_train, y_train);
```

```python
y_train_pred = model.predict(x_train);
y_test_pred = model.predict(x_test);
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mse_train = mean_squared_error(y_train, y_train_pred);
mae_train = mean_squared_error(y_train, y_train_pred);
r2_train= r2_score(y_train, y_train_pred);
```

```python
mse_test=mean_squared_error(y_test,y_test_pred)
mae_test=mean_squared_error(y_test,y_test_pred)
r2_test = r2_score(y_test, y_test_pred)
```
