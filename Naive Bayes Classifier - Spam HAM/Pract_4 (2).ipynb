```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/spam.csv',encoding='ISO-8859-1')
df.head()
```

```python
df.groupby('v1').describe()
```

```python
df['spam'] = df['v1'].apply(lambda x:1 if x=='spam' else 0)
```

```python
df.head()
```

```python
new_df = df[['v1','v2','spam']]
```

```python
new_df.head()
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(df.v2,df.spam)
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer()
x_train_count=v.fit_transform(x_train.values)
x_train_count.toarray()[:2]
```

```python
print(v)
print(df.v2)
print(df.spam)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train_count,y_train)
```

```python
from sklearn.naive_bayes import BernoulliNB
model1=BernoulliNB()
model1.fit(x_train_count,y_train)
```

```python
from sklearn.naive_bayes import GaussianNB
model2=GaussianNB()
model2.fit(x_train_count.toarray(),y_train)
```

```python
emails=["Hello Buddy, what's up!", "Party tonight?"]

email_count=v.transform(emails)
model.predict(email_count)
```

```python
x_test_count=v.transform(x_test)
model.score(x_test_count,y_test)
```

```python
emails=["Hello Buddy, what's up!", "Party tonight?"]

email_count=v.transform(emails)
model1.predict(email_count)
```

```python
x_test_count=v.transform(x_test)
model1.score(x_test_count,y_test)
```
